```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')

dataframe = pd.read_csv('/content/drive/MyDrive/ArquivosColab/titles.csv')
```

Tabulaçao Cruzada

```python
pd.crosstab(dataframe.production_countries, dataframe.type)
```

Gerar Total

```python
pd.crosstab(dataframe.production_countries, dataframe.type, margins=True)
```

Proporções

```python
pd.crosstab(dataframe.production_countries, dataframe.type, margins=True, normalize='columns')*100
```

Inserir Terceira variável

```python
pd.crosstab(dataframe.production_countries, [dataframe.type, dataframe.imdb_score.round(0)], margins=True, normalize='columns')*100
```

Inserir Médias na Tabulação Cruzada

```python
import numpy as np

pd.crosstab(dataframe.production_countries, dataframe.type, values= dataframe.runtime, aggfunc = np.average, margins=True)
```
