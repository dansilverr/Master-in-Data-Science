```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')

dataframe = pd.read_csv('/content/drive/MyDrive/ArquivosColab/titles.csv')
```

Encontrar Medidas nos Dados com describe()

```python
dataframe[['imdb_score', 'tmdb_popularity']].describe().round(2)
```

Calcular Medidas com GroupBy

```python
dataframe.groupby('production_countries')[['tmdb_popularity','imdb_score']].mean()

```

```python
dataframe.groupby('production_countries')[['tmdb_popularity','imdb_score']].median()
```

```python
dataframe.groupby('production_countries')[['tmdb_popularity','imdb_score']].var()
```

```python
dataframe.groupby('production_countries')[['tmdb_popularity','imdb_score']].std()
```

```python
dataframe.groupby('production_countries')[['tmdb_popularity','imdb_score']].quantile(0.25)
```

BoxPlots

```python
import seaborn as sns

sns.boxplot(y = 'imdb_score', x = 'production_countries', data = dataframe)
```
